```python
import pandas as pd
import numpy as np 
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df=pd.read_csv('train_and_test2.csv')
df
```

```python
#Make passenger ID as index
df1=df.set_index('Passengerid')
df1
```

```python
#To delete columns with all zeros
df2 = df1.loc[:, (df != 0).any(axis=0)]
df2
```

```python
#Delete rows with missing values
df3 = df2[~np.isnan(df2).any(axis=1)]
df3
```

```python
# Summary statistics 
display(df3.describe())
```

```python
# Mean, Median & Mode
meanData = df3.mean()
meanData
```

```python
medianData = df3.median()
medianData
```

```python
modeData = df3.mode()
modeData
```

```python
df3.hist(figsize=(10,10),bins=10)
```


## From the above data analysis and visualizations it can be concluded that
### Sex, Pclass, Embarked, Survived are categorical variables. 
### Standard deviation is more for age and fare that indicate data is widely distributed from mean. 
### Mean, Median, Mode is almost same for Sex variable hence distribution will be more close to normal distribution. 
### Mean, Median & Mode for parameters Fare, sibsp, Parch have large difference. Hence these distributions will be more skewed.




```python
# Correlation matrix
corrMatrix = df3.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

### No pair of parameters have the strong correlation. 
### Highest correlation is -0.56 between Pclass and Fare. 
### 2nd highest correlation is -0.38 between Pclass and Age.

Automatic driving car without human intervention:

Complex coding tasks like auto driving, gaming etc. can be modelled by reinforcement learning method without being explicitly coded. One more advantage for this model is no previous test data required. 

Object learns from actions performed in an environment based on reward received. 

For the example of driving, car is the object. Environment is road, signals, other vehicles around and instructions from maps etc.

If it perform wrong action, based on the mistake proportionate reward function will cut down. 

With this trail and error actions model trains and it remembers good actions where it got highest reward function and always tries to repeat them.
